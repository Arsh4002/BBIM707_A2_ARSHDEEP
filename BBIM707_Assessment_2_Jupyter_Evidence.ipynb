```python
import pandas as pd
import numpy as np
```

```python
retail_df = pd.read_csv("Online Retail.csv", encoding='ISO-8859-1')

retail_df.head()
```

```python
print(retail_df.shape)
```

```python
customer_df = pd.read_csv("marketing_campaign.csv")

customer_df.head()
```

```python
print(customer_df.shape)
```

```python
print(retail_df.isnull().sum())
```

```python
retail_df = retail_df.dropna()
```

```python
print(customer_df.isnull().sum())
```

```python
customer_df["Income"] = customer_df["Income"].fillna(
    customer_df["Income"].median()
)
```

```python
print(retail_df.isnull().sum())
```

```python
print(customer_df.isnull().sum())
```

```python
print(
    "Duplicate Rows:",
    retail_df.duplicated().sum()
)

retail_df = retail_df.drop_duplicates()
```

```python
print(
    "Duplicate Rows:",
    customer_df.duplicated().sum()
)

customer_df = customer_df.drop_duplicates()
```

```python
print(retail_df.shape)
```

```python
print(customer_df.shape)
```

```python
customer_df.rename(
    columns={
        'ID':'CustomerID',
        'Year_Birth':'BirthYear'
    },
    inplace=True
)
```

```python
print(customer_df.columns)
```

```python
print(retail_df.dtypes)
```

```python
print(customer_df.dtypes)
```

```python
retail_df['InvoiceDate'] = pd.to_datetime(
    retail_df['InvoiceDate'],
    dayfirst=True
)
```

```python
customer_df['Dt_Customer'] = pd.to_datetime(
    customer_df['Dt_Customer'],
    dayfirst=True
)
```

```python
print(retail_df.dtypes)
```

```python
print(customer_df.dtypes)
```

```python
retail_df['TotalSales'] = (
    retail_df['Quantity']
    *
    retail_df['UnitPrice']
)
```

```python
customer_df['Age'] = (
    2025 -
    customer_df['BirthYear']
)
```

```python
customer_df['AgeGroup'] = pd.cut(
    customer_df['Age'],
    bins=[0,30,45,60,100],
    labels=[
        'Young',
        'Adult',
        'Middle Age',
        'Senior'
    ]
)
```

```python
print(
    retail_df[
        ['Quantity',
         'UnitPrice',
         'TotalSales']
    ].head()
)
```

```python
print(
    customer_df[
        ['Age',
         'AgeGroup']
    ].head()
)
```

```python
customer_sales = retail_df.groupby(
    'CustomerID'
).agg({
    'TotalSales':'sum',
    'Quantity':'sum',
    'InvoiceNo':'nunique'
}).reset_index()

customer_sales.rename(
    columns={
        'InvoiceNo':'NumOrders'
    },
    inplace=True
)

print(customer_sales.head())
```

```python
min_rows = min(
    len(customer_sales),
    len(customer_df)
)

integrated_df = pd.concat(
    [
        customer_sales.iloc[:min_rows]
        .reset_index(drop=True),

        customer_df.iloc[:min_rows]
        .reset_index(drop=True)
    ],
    axis=1
)
```

```python
integrated_df.head()
```

```python
print(integrated_df.shape)
```

```python
integrated_df.to_csv(
    "Integrated_Customer_Data.csv",
    index=False
)

print(
    "Integrated dataset saved successfully."
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
integrated_df = pd.read_csv(
    "Integrated_Customer_Data.csv"
)

integrated_df.head()
```

```python
summary_stats = integrated_df.describe()
summary_stats
```

```python
print(
    integrated_df['AgeGroup']
    .value_counts()
)
```

```python
print(
    integrated_df['Education']
    .value_counts()
)
```

```python
print(
    integrated_df['Marital_Status']
    .value_counts()
)
```

```python
sales_age = integrated_df.groupby(
    'AgeGroup'
)['TotalSales'].mean()

sales_age.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Average Total Sales by Age Group'
)

plt.xlabel('Age Group')
plt.ylabel('Average Total Sales')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

n, bins, patches = plt.hist(
    integrated_df['Income'],
    bins=20
)

# Add frequency labels on top of bars
for i in range(len(patches)):
    height = patches[i].get_height()
    if height > 0:
        plt.text(
            patches[i].get_x() + patches[i].get_width()/2,
            height,
            int(height),
            ha='center',
            va='bottom'
        )

plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
sales_education = integrated_df.groupby(
    'Education'
)['TotalSales'].mean()

sales_education.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title(
    'Average Sales by Education Level'
)

plt.xlabel('Education')
plt.ylabel('Average Total Sales')

plt.tight_layout()
plt.show()
```
