```python
import pandas as pd

df = pd.read_csv(
    "/content/Integrated_Customer_Data.csv"
)

df.head()
```

```python
import json
import time

records = df.head(20)

for index, row in records.iterrows():

    message = row.to_dict()

    print(
        "Sending:",
        json.dumps(message)
    )

    time.sleep(1)
```

```python
stream_messages = []

for index, row in records.iterrows():

    stream_messages.append(
        row.to_dict()
    )

print(
    "Messages in Stream:",
    len(stream_messages)
)
```

```python
for message in stream_messages:

    print(
        "Received:",
        message
    )
```

```python
running_total = 0

for message in stream_messages:

    running_total += message["TotalSales"]

    print(
        "Current Sales Total:",
        running_total
    )
```

```python
stream_df = pd.DataFrame(
    stream_messages
)

stream_df.head()
```
